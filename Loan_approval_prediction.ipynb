```python
import pandas as pd
```

```python
data = pd.read_csv("loan_approval_dataset.csv")
```

```python
data.drop(columns= ['loan_id'],inplace=True)
```

```python
data.columns = data.columns.str.strip()
```

```python
data.columns
```

```python
data['Assets'] = data.residential_assets_value + data.bank_asset_value + data.commercial_assets_value + data.luxury_assets_value
```

```python
data
```

```python
data.drop(columns = ['residential_assets_value',	'commercial_assets_value',	'luxury_assets_value',	'bank_asset_value'], inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.education.unique()
```

```python
def clean_data(st):
  st = st.strip()
  return st
data.education = data.education.apply(clean_data)
```

```python
data.education.unique()
```

```python
data['education'] = data['education'].replace(['Graduate','Not Graduate'],[1,0])
```

```python
data
```

```python
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.apply(clean_data)
```

```python
data.self_employed.unique()
```

```python
data['self_employed'] = data['self_employed'].replace(['Yes','No'],[1,0])
```

```python
data
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.apply(clean_data)
```

```python
data.loan_status.unique()
```

```python
data['loan_status'] = data.loan_status.replace(['Approved','Rejected'],[1,0])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['loan_status'],axis=1)
Y = data['loan_status']
X.shape,Y.shape

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.4,
                                                    random_state=1)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3)
rfc = RandomForestClassifier(n_estimators = 7,
                             criterion = 'entropy',
                             random_state =7)
svc = SVC()
lc = LogisticRegression()

# making predictions on the training set
for clf in (rfc, knn, svc,lc):
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_train)
    print("Accuracy score of ",
          clf.__class__.__name__,
          "=",100*metrics.accuracy_score(Y_train,
                                         Y_pred))
```

```python
import joblib
joblib.dump(rfc, 'loan_model.pkl')
```
